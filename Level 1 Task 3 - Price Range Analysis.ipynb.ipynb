```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset (internship).csv")
df.head()
```

```python
df['Price range'].unique()
```

```python
price_counts = df['Price range'].value_counts().sort_index()
price_counts
```

```python
price_counts.plot(kind='bar', figsize=(6,4))
plt.title("Restaurant Distribution by Price Range")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
most_common_price = price_counts.idxmax()
print("Most common price range is:", most_common_price)
```

Level 1 – Task 3: Price Range Analysis

In this task, I analyzed the distribution of restaurants across different price ranges. I calculated the frequency of restaurants in each price range and visualized the distribution using a bar chart. From the analysis, I identified the most common price range, which represents the price category with the highest number of restaurants.
