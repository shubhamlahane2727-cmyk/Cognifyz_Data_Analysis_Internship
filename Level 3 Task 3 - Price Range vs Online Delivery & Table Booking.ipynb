```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset (internship).csv")
df.head()
```

```python
df['Price range'].value_counts().sort_index()
```

```python
online_delivery_by_price = (
    df.groupby('Price range')['Has Online delivery']
    .value_counts(normalize=True)
    .unstack()
)

online_delivery_by_price
```

```python
online_delivery_by_price['Yes'].plot(kind='bar', figsize=(7,5))
plt.title("Online Delivery Availability by Price Range")
plt.xlabel("Price Range")
plt.ylabel("Proportion of Restaurants")
plt.xticks(rotation=0)
plt.show()
```

```python
table_booking_by_price = (
    df.groupby('Price range')['Has Table booking']
    .value_counts(normalize=True)
    .unstack()
)

table_booking_by_price
```

```python
table_booking_by_price['Yes'].plot(kind='bar', figsize=(7,5))
plt.title("Table Booking Availability by Price Range")
plt.xlabel("Price Range")
plt.ylabel("Proportion of Restaurants")
plt.xticks(rotation=0)
plt.show()
```

Level 3 Task 3 - Price Range vs Online Delivery & Table Booking
In this task, I analyzed the relationship between restaurant price range and the availability of online delivery and table booking services. The proportion of restaurants offering these services was calculated for each price range. The results show that higher-priced restaurants tend to offer table booking more frequently, while online delivery is more common among mid-range restaurants. This suggests that service availability varies based on pricing strategy and customer expectations.
