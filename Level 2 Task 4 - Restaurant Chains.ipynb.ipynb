```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset (internship).csv")
df.head()
```

```python
chain_counts = df['Restaurant Name'].value_counts()
chains = chain_counts[chain_counts > 1]
chains.head(10)
```

```python
restaurant_chains = df['Restaurant Name'].value_counts()
restaurant_chains = restaurant_chains[restaurant_chains > 1]
```

```python
chain_analysis = df[df['Restaurant Name'].isin(restaurant_chains.index)].groupby(
    'Restaurant Name'
).agg(
    Outlets=('Restaurant Name', 'count'),
    Avg_Rating=('Aggregate rating', 'mean'),
    Total_Votes=('Votes', 'sum')
).sort_values(by='Outlets', ascending=False)

chain_analysis.head(10)
```

```python
chain_analysis.head(10)['Outlets'].plot(kind='bar', figsize=(9,5))
plt.title("Top 10 Restaurant Chains by Number of Outlets")
plt.xlabel("Restaurant Chain")
plt.ylabel("Number of Outlets")
plt.xticks(rotation=45, ha='right')
plt.show()
```
