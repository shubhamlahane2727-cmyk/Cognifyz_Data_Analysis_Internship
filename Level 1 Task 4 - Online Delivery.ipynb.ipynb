```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset (internship).csv")
df.head()
```

```python
online_delivery_counts = df['Has Online delivery'].value_counts()
online_delivery_counts
```

```python
online_delivery_counts.plot(kind='bar', figsize=(5,4))
plt.title("Online Delivery Availability")
plt.xlabel("Online Delivery")
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
most_common = online_delivery_counts.idxmax()
print("Most restaurants have online delivery:", most_common)
```

Level 1 – Task 4: Online Delivery Analysis

In this task, I analyzed the availability of online delivery services among restaurants. I calculated the number of restaurants that offer online delivery and those that do not. A bar chart was used to visualize the distribution. From the analysis, it was observed whether online delivery is commonly available among restaurants.
