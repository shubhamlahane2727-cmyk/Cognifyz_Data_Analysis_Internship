```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset (internship).csv")
df.head()
```

```python
df['Aggregate rating'].describe()
```

```python
plt.figure(figsize=(7,5))
plt.hist(df['Aggregate rating'], bins=10)
plt.xlabel("Aggregate Rating")
plt.ylabel("Number of Restaurants")
plt.title("Distribution of Restaurant Ratings")
plt.show()
```

```python
rating_bins = pd.cut(
    df['Aggregate rating'],
    bins=[0,1,2,3,4,5]
)

rating_bins.value_counts().sort_index()
```

```python
average_votes = df['Votes'].mean()
average_votes
```

```python
print("Average number of votes received by restaurants:", round(average_votes, 2))
```
