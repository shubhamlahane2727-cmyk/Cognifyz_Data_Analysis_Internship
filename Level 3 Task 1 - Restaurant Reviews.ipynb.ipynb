```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset (internship).csv")
df.head()
```

```python
df[['Votes', 'Aggregate rating']].describe()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Votes'], df['Aggregate rating'], alpha=0.5)
plt.xlabel("Number of Votes")
plt.ylabel("Aggregate Rating")
plt.title("Relationship Between Votes and Ratings")
plt.show()
```

```python
correlation = df['Votes'].corr(df['Aggregate rating'])
print("Correlation between Votes and Ratings:", round(correlation, 3))
```

Level 3 – Task 1: Restaurant Reviews
In this task, I analyzed the relationship between the number of votes and aggregate ratings of restaurants. A scatter plot was used to visualize how ratings vary with customer engagement. The correlation value indicates the strength of the relationship between votes and ratings. This analysis helps understand whether popular restaurants tend to receive higher ratings.
