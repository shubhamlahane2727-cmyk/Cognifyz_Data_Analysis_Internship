```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset (internship).csv")
df.head()
```

```python
df['Votes'].describe()
```

```python
plt.figure(figsize=(7,5))
plt.hist(df['Votes'], bins=20)
plt.xlabel("Number of Votes")
plt.ylabel("Number of Restaurants")
plt.title("Distribution of Votes")
plt.show()
```

```python
top_voted = df.sort_values(by='Votes', ascending=False)[
    ['Restaurant Name', 'Votes', 'Aggregate rating']
].head(10)

top_voted
```

```python
top_voted.set_index('Restaurant Name')['Votes'].plot(
    kind='bar', figsize=(9,5)
)
plt.title("Top 10 Restaurants by Votes")
plt.xlabel("Restaurant Name")
plt.ylabel("Votes")
plt.xticks(rotation=45, ha='right')
plt.show()
```

Level 3 – Task 2: Votes Analysis

In this task, I analyzed the distribution of customer votes to understand engagement levels across restaurants. A histogram was used to visualize how votes are distributed, and the top restaurants with the highest number of votes were identified. This analysis highlights the popularity of restaurants and customer participation patterns.
