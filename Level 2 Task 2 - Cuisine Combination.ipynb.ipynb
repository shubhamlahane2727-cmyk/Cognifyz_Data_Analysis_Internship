```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset (internship).csv")
df.head()
```

```python
cuisine_combo_counts = df['Cuisines'].value_counts().head(10)
cuisine_combo_counts
```

```python
combo_avg_rating = (
    df.groupby('Cuisines')['Aggregate rating']
    .mean()
    .sort_values(ascending=False)
)

combo_avg_rating.head(10)
```

```python
combo_avg_rating.head(10).plot(kind='bar', figsize=(9,5))
plt.title("Top 10 Cuisine Combinations by Average Rating")
plt.xlabel("Cuisine Combination")
plt.ylabel("Average Rating")
plt.xticks(rotation=45, ha='right')
plt.show()
```

Level 2 – Task 2: Cuisine Combination Analysis

In this task, I analyzed the combinations of cuisines served by restaurants to identify the most frequently occurring cuisine combinations. I also calculated the average aggregate rating for each cuisine combination to determine whether certain combinations tend to receive higher ratings. This analysis helps in understanding customer preferences for specific cuisine pairings.
